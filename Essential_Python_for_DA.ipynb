## OOP

```python
class Dog:
  def __init__(self,name,species, age):
    self.name = name
    self.species = species
    self.age = age

  def bark(self):
    print(f"{self.name} says woof!")

  def hbd(self):
    self.age += 1
```

```python
dog1 = Dog("milo","Chihuahua",2)
dog2 = Dog("david","Husky",3)
```

```python
print(f"{dog1.name} is {dog1.species}")
```

```python
print(f"{dog2.name} is {dog2.species}")
```

```python
dog1.bark()
```

```python
dog1.hbd()
```

```python
print(dog1.age)
```

```python
class ATM:
  def __init__(self, username, bank, amount):
    self.username = username
    self.bank = bank
    self.amount = amount

  def __str__(self):
    return f"{self.username} uses {self.bank} and has {self.amount} THB in an account."

  def check_balance(self):
    print(f"Your balance is {self.amount} THB")

  def deposit(self, saving):
    self.amount += saving
    print(f"Deposit {saving} THB successfully")
```

```python
printer = ATM("sarunpron","scb",100)
```

```python
printer.check_balance()
```

```python
printer.deposit(100)
```

```python
print(printer.amount)
```

```python
print(printer)
```

```python
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age

## extension
class Employee(Person):
  def __init__(self,name,age,company):
    super().__init__(name,age)   ## refer to class Person
    self.company = company

  def greeting(self):
    print("Hello World!")
```

```python
person1 = Person("Printer", 27)
```

```python
person2 = Employee("Lisa", 28, "google")
```

```python
person2.greeting()
```

```python
## read file
import csv
```

```python
try:
  file = open("hotel.csv","r") ## r w a

  data = csv.reader(file)

  for row in data:
    print(row)

  file.close()

except FileNotFoundError:
  print("Error, check again.")
finally:
  print("read file successfully.")
```

```python
# context manager
read_data = [] ## empty list

with open("hotel.csv", "r") as file:
  data = csv.reader(file)
  for row in data:
    read_data.append(row)
```

```python
# header
read_data[0]
```

```python
# body data
read_data[1:]
```

```python
## for loop to calculate avg. price per night
prices = []

for row in read_data[1: ]:
   prices.append(int(row[-1]))  # เก็บ row ที่ index -1 ไปไว้ใน prices

print(f"Average price per night: {sum(prices)/len(prices)}")
```

```python
## for loop
import requests
import time

charecters = []

for i in range(5):
    url = f"https://swapi.info/api/people/{i+1}"
    response = requests.get(url)
    response_json = response.json()
    charecters.append(response_json["name"])
    print(f"Success: {i+1}")
    time.sleep(0.5)

print(charecters)
```
