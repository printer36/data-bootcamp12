## Control Flow
*   if

```python
score = 90

if score >= 80:
  print("passed")
else:
  print("failed")
```

```python
score = 65

if score >= 80:
  print("A")
elif score >= 70:
  print("B")
elif score >= 60:
  print("C")
else:
  print("D")
```
