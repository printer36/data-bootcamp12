```python
## map / filter มักใช้คู่กับ lambda
## iterable (adj)
scores = [82, 78, 81, 66, 50]
```

```python
def grading(score):
  if score >= 80:
    return("passed")
  else:
    return("failed")
```

```python
## map
list(map(grading,scores))
```

```python
## add five to scores (แบบ for loop)
new_scores = []
for score in scores:
  new_scores.append(score+5)
print(new_scores)
```

```python
## add five to scores (แบบ lambda)
list(map(lambda x: x+5,scores))
```

```python
## filter
scores
```

```python
list(filter(lambda x: x>=80, scores))
```

```python
## เพิ่มชื่อนักเรียน
scores = [("toy", 80), ("jane", 78), ("lisa", 95)]

## filter score>=80
list(filter(lambda x: x[1]>=80, scores))
```

```python
## module/ library
## for homework class ATM: add OTP
import random

class ATM:
  def __init__(self, name, balance):
    self.name = name
    self.balance = balance

  def __str__(self):
    return f"Account name: {self.name}"

  def get_opt(self):
    otp = random.randint(10000,99999)
    return f"Your Requested OTP: {otp}"
```

```python
my_account = ATM("Printer", 500)
```

```python
print(my_account)
my_account.get_opt()
```
